# Fuel economy between different car manufacturers overtime

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

car_df =  pd.read_csv('Dataset.csv')
```

### 1. Find the quantity of car models in the period of 2015 - 2023.

```python
df = car_df.groupby(['Model Year'])['Division'].nunique().plot()
plt.xlabel("Year")
plt.ylabel("# Car Models")
plt.title("Number of Car Models per year")
plt.show()
```

### 2a. Find the quantity of car models for Aston Martin each year.

### 2b. Find the quantity of car models for BMW each year.

### 2c. Find the quantity of car models for Audi each year.

### 2d. Find the quantity of car models for Lamborghini each year.

### 2e. Find the quantity of car models for Acura each year.

```python
astondf = car_df[car_df['Mfr Name'] == 'Aston Martin']
astondf.groupby(['Model Year'])['Carline'].nunique().plot()

bmwdf = car_df[car_df['Mfr Name'] == 'BMW']
bmwdf.groupby(['Model Year'])['Carline'].nunique().plot()

audidf = car_df[car_df['Division'] == 'Audi']
audidf.groupby(['Model Year'])['Carline'].nunique().plot()

lamdf = car_df[car_df['Division'] == 'Lamborghini']
lamdf.groupby(['Model Year'])['Carline'].nunique().plot()

acuradf = car_df[car_df['Division'] == 'Acura']
acuradf.groupby(['Model Year'])['Carline'].nunique().plot()
plt.xlabel("Year")
plt.ylabel("# Car Models")
plt.title("Quantity of Models each year")
plt.legend(['Aston Martin', 'BMW', 'Audi', 'Lamborghini', 'Acura'])
plt.show()
```

### 3. Find the car models with the best and worst FE ratings each year.

```python
years = car_df['Model Year'].unique()
FE = {}
for year in years:
    yeardf = car_df[car_df['Model Year'] == year]
    topFE = yeardf['Combined FE'].max()
    lowFE = yeardf['Combined FE'].min()
    topname=yeardf[yeardf['Combined FE'] == topFE]['Division'].values[0]
    lowname=yeardf[yeardf['Combined FE'] == lowFE]['Division'].values[0]
    FE[year] = {"Best FE": (topname, topFE), "Worst FE": (lowname, lowFE)}
print(FE)
```

### 4a. Find the FE rating for Bugatti each year.

### 4b. Find the FE rating for Hyundai Motor Company each year.

### 4c. Find the FE rating for Mercedes-Benz each year.

### 4d. Find the FE rating for Toyota each year.

### 4e. Find the FE rating for Ferrari North America, Inc. each year.

```python
car_df['average FE'] = car_df[['City FE', 'Highway FE']].mean(axis=1)

bgtdf = car_df[car_df['Division'] == 'Bugatti']
bgtdf.groupby(['Model Year'])['average FE'].mean().plot()

hmcdf = car_df[car_df['Division'] == 'Hyundai Motor Company']
hmcdf.groupby(['Model Year'])['average FE'].mean().plot()

mercdf = car_df[car_df['Division'] == 'Mercedes-Benz']
mercdf.groupby(['Model Year'])['average FE'].mean().plot()

toydf = car_df[car_df['Division'] == 'Toyota']
toydf.groupby(['Model Year'])['average FE'].mean().plot()

ferdf = car_df[car_df['Division'] == 'Ferrari North America, Inc.']
ferdf.groupby(['Model Year'])['average FE'].mean().plot()

plt.xlabel("Year")
plt.ylabel("Average FE Rating")
plt.title("FE rating each year")
plt.legend(['Bugatti', 'Hyundai Motor Company', 'Mercedes-Benz', 'Toyota', 'Ferrari North America, Inc.'])
plt.show()
```
